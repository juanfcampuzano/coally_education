```python
from pymongo import MongoClient
from langchain.embeddings.openai import OpenAIEmbeddings
from langchain.vectorstores import MongoDBAtlasVectorSearch
from langchain.document_loaders import DirectoryLoader
from langchain.llms import OpenAI
from langchain.chains import RetrievalQA
from langchain_core.documents.base import Document
from pymongo import MongoClient
import os
import pandas as pd
```

```python
client = MongoClient(os.getenv("MONGO_URI"))
dbName = os.getenv("EMBEDDINGS_DB_NAME")
collectionName = os.getenv("EMBEDDINGS_COLLECION_NAME")
collection = client[dbName][collectionName]
```

```python
df = pd.read_csv("cursos_general.csv")
docs = []
for index, row in df.iterrows():
    metadata = {"id":row["id"], "titulo":row["titulo"], "url":row["url"]}
    page_content = str(row["titulo"]) + " " + str(row["descripcion"])
    document = Document(metadata=metadata, page_content=page_content)
    docs.append(document)
```

```python
embeddings = OpenAIEmbeddings(openai_api_key=os.getenv("OPENAI_API_KEY"))

vectorStore = MongoDBAtlasVectorSearch.from_documents( docs, embeddings, collection=collection )
```

```python
embeddings = OpenAIEmbeddings(openai_api_key=os.getenv("OPENAI_API_KEY"))
vectorStore = MongoDBAtlasVectorSearch( collection, embeddings )
```

```python
def query_data(query): 
    docs = vectorStore.similarity_search(query, k=5)
    return docs
```
